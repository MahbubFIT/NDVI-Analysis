```python
# ============================================================
# 02_LST_trend_hotspot_analysis.ipynb
# ============================================================
# Land Surface Temperature (LST) preprocessing, zonal statistics,
# seasonal diagnostics, and hotspot/coolspot analysis for Dhaka.
#
# This notebook implements a complete workflow for:
#   • Landsat 8/9 LST extraction (2014–2024)
#   • Cloud masking, radiometric calibration, emissivity correction
#   • Single‑channel LST retrieval following Jiménez‑Muñoz et al. (2008, 2014)
#   • Multi‑month LST mosaics and zonal statistics for Dhaka wards
#   • Monthly and seasonal LST mapping (Jan–May, Oct–Dec)
#   • Hotspot and coolspot identification
#   • ΔLST (April–December) thermal contrast analysis
#   • K‑means clustering of monthly and seasonal heat‑risk patterns
#   • MODIS LST integration for long‑term consistency checks
#
# Long‑term LST trends are assessed using two complementary
# non‑parametric methods: the Mann–Kendall (MK) test for detecting
# monotonic trends, and the Theil–Sen estimator for quantifying the
# rate of change (°C/year). This combination provides a robust,
# distribution‑free assessment of thermal trajectories across Dhaka.
#
# The workflow is informed by established thermal remote‑sensing
# practices and recent applications in urban heat research, including:
#   • Mohiuddin & Mund (2024) – Dhaka LST preprocessing framework
#   • Jiménez‑Muñoz et al. (2008, 2014) – Single‑channel LST algorithm
#
# Dhaka‑specific parameterization, preprocessing logic, zonal extraction,
# visualization design, and all implementation decisions were developed
# by Mahbubul Alam.
#
# This notebook supports the LST trend, hotspot persistence, and
# vulnerability diagnostics presented in the thesis.
# ============================================================


```

```python
######Zone-wise Monthly Land Surface Temperature (2017) – Dhaka#####

# -----------------------------
# Paths
# -----------------------------
folder = "/content/drive/MyDrive/GEE_LST_Dhaka_2017"
shp_path = "/content/drive/MyDrive/ARS/Coredhaka.shp"

# -----------------------------
# Load zones (KEEP WGS84)
# -----------------------------
zones = gpd.read_file(shp_path)   # DO NOT reproject
xmin, ymin, xmax, ymax = zones.total_bounds

# -----------------------------
# List raster files
# -----------------------------
files = sorted([f for f in os.listdir(folder) if f.endswith(".tif")])

month_names = ["Jan", "Feb", "Mar", "Apr",
               "May", "Jun", "Jul", "Aug",
               "Sep", "Oct", "Nov", "Dec"]

exclude_months = ["Jun", "Jul", "Aug", "Sep"]

vmin, vmax = 20, 45

zone_maps = []
zone_months = []

# -----------------------------
# Zonal statistics (WGS84-safe)
# -----------------------------
for fname, month in zip(files, month_names):

    if month in exclude_months:
        continue

    path = os.path.join(folder, fname)

    stats = zonal_stats(
        zones,          # WGS84 zones
        path,           # WGS84 raster
        stats=["mean"],
        geojson_out=True,
        nodata=np.nan
    )

    gdf = gpd.GeoDataFrame.from_features(stats)
    gdf["mean"] = gdf["mean"].astype(float)

    zone_maps.append(gdf)
    zone_months.append(month)

# -----------------------------
# Plot zone-wise maps
# -----------------------------
n = len(zone_maps)
cols = 4
rows = int(np.ceil(n / cols))

fig, axes = plt.subplots(rows, cols, figsize=(14, 3.5 * rows))
axes = np.array(axes).reshape(-1)
used_axes = axes[:n]

# Reduced bottom margin to bring maps closer to the figure bottom
# Adjusted bottom to give less space, bringing maps closer to the figure bottom
fig.subplots_adjust(bottom=0.2, top=0.9, left=0.05, right=0.95,
                    hspace=0.1, wspace=0.05)

for ax, gdf, month in zip(used_axes, zone_maps, zone_months):

    gdf.plot(
        column="mean",
        cmap="RdYlGn_r",
        vmin=vmin,
        vmax=vmax,
        linewidth=0.3,
        edgecolor="black",
        ax=ax,
        legend=False
    )

    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    ax.set_aspect("equal")
    ax.set_title(month, fontsize=10)
    ax.axis("off")

for ax in axes[n:]:
    ax.axis("off")

# ---------------------------------------------------------
# Unified Legend Panel
# ---------------------------------------------------------
# Adjusted bottom to be even lower (closer to figure bottom) and height to be smaller
legend_ax = fig.add_axes([0.1, 0.02, 0.8, 0.12]) # [left, bottom, width, height]
legend_ax.axis("off")

# Define a common vertical position for elements in legend_ax
common_y_pos = 0.5

# ---------------------------------------------------------
# 1. North Arrow
# ---------------------------------------------------------
na_inset_ax = legend_ax.inset_axes([0.02, common_y_pos - 0.25, 0.08, 0.5]) # [x, y, width, height] relative to legend_ax
na_inset_ax.axis("off")

# Draw the arrow
from matplotlib.patches import FancyArrowPatch # Ensure FancyArrowPatch is imported
arrow = FancyArrowPatch(
    posA=(0.5, 0.2),  # Tail of the arrow (moved down)
    posB=(0.5, 0.9),  # Head of the arrow (moved up for length)
    mutation_scale=20, # Increased size of the arrowhead
    color='black',
    arrowstyle='-|>', # Standard arrow style
    linewidth=2,       # Increased line thickness
    transform=na_inset_ax.transAxes # Coords are axes fraction
)
na_inset_ax.add_patch(arrow)

# Place 'N' text above the arrow
na_inset_ax.text(
    0.5, 0.95, # Position the 'N' higher up than the arrow head
    "N",
    ha="center",
    va="center",
    fontsize=12, # Increased fontsize
    fontweight="bold",
    transform=na_inset_ax.transAxes
)

# ---------------------------------------------------------
# 2. LST Colorbar
# ---------------------------------------------------------
sm = plt.cm.ScalarMappable(
    cmap="RdYlGn_r",
    norm=plt.Normalize(vmin=vmin, vmax=vmax)
)
sm._A = []

# Slim, elegant colorbar, positioned more centrally, reduced width and height
cax = legend_ax.inset_axes([0.2, common_y_pos - 0.1, 0.4, 0.2])  # [x, y, width, height] relative to legend_ax
cbar = fig.colorbar(sm, cax=cax, orientation="horizontal")

# Make it slim and clean
cbar.outline.set_linewidth(0.4)
cbar.ax.tick_params(labelsize=7, width=0.4, length=2)
cbar.set_label("LST (°C)", fontsize=8, labelpad=2)


# ---------------------------------------------------------
# 3. Scale Bar (WGS84)
# ---------------------------------------------------------
# Dhaka latitude ≈ 23.8°
meters_per_degree_lat = 110574  # constant
deg_for_2km = 2000 / meters_per_degree_lat

scale_y = common_y_pos # Aligned vertically with other elements
scale_x_start = 0.75 # Positioned right
scale_width_rel = 0.12  # fixed visual width

legend_ax.plot(
    [scale_x_start, scale_x_start + scale_width_rel],
    [scale_y, scale_y],
    transform=legend_ax.transAxes,
    color="black",
    linewidth=4
)

legend_ax.text(
    scale_x_start + scale_width_rel / 2,
    scale_y - 0.15, # Adjusted text position
    "2 km",
    ha="center",
    va="top",
    fontsize=10,
    transform=legend_ax.transAxes
)

# -----------------------------
# Title
# -----------------------------
fig.suptitle(
    "Zone-wise Monthly Land Surface Temperature (2017) – Dhaka\n"
    "June, July, August & September excluded due to cloud contamination",
    fontsize=10,
    y=0.97
)

plt.show()
```

```python
#### Seasonal LST variations ####

# ---------------------------------------------------------
# Load shapefile
# ---------------------------------------------------------
shp_path = "/content/drive/MyDrive/ARS/Coredhaka.shp"
zones = gpd.read_file(shp_path)

# Month mapping
month_map = {
    "01": "Jan", "02": "Feb", "03": "Mar", "04": "Apr",
    "05": "May", "06": "Jun", "07": "Jul", "08": "Aug",
    "09": "Sep", "10": "Oct", "11": "Nov", "12": "Dec"
}

# Months to plot (including Jan and May)
selected_months = ["Jan", "Feb", "Mar", "Apr", "May", "Oct", "Nov", "Dec"]

# ---------------------------------------------------------
# Load LST zonal statistics
# ---------------------------------------------------------
folder = "/content/drive/MyDrive/GEE_LST_Dhaka_2017"

files = sorted([
    f for f in os.listdir(folder)
    if f.startswith("LST_Dhaka_2017_") and f.endswith(".tif")
])

lst_values = {}

for f in files:
    month_num = f.split("_")[-1].split(".")[0]
    month_name = month_map.get(month_num)

    if month_name not in selected_months:
        continue

    stats = zonal_stats(zones, os.path.join(folder, f), stats=["mean"], nodata=np.nan)
    lst_values[month_name] = np.array([s["mean"] for s in stats])

# Print months with loaded data to check for 'May'
print("Months with loaded LST values:", lst_values.keys())

# ---------------------------------------------------------
# Detect zones with >5°C change
# ---------------------------------------------------------
threshold = 5  # degrees Celsius

# Build matrix: zones × months
matrix = np.vstack([lst_values[m] for m in selected_months]).T

# Compute max difference per zone
max_diff_per_zone = np.nanmax(matrix, axis=1) - np.nanmin(matrix, axis=1)

# Zones to label
significant_zones = np.where(max_diff_per_zone > threshold)[0]

# ---------------------------------------------------------
# Hotspot (April max) and Coolspot (November min)
# ---------------------------------------------------------
hot_month = "Apr"
cool_month = "Nov"

hot_idx = np.nanargmax(lst_values[hot_month])
cool_idx = np.nanargmin(lst_values[cool_month])

hot_pt = zones.geometry[hot_idx].centroid
cool_pt = zones.geometry[cool_idx].centroid

hot_name = zones.loc[hot_idx, "NAME_3"]
cool_name = zones.loc[cool_idx, "NAME_3"]

# ---------------------------------------------------------
# Create figure
# ---------------------------------------------------------
# Adjusted for 2 rows x 4 columns for 8 months
fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.flatten()

vmin, vmax = 20, 45

fig.subplots_adjust(bottom=0.25, top=0.9, left=0.05, right=0.95,
                    hspace=0.1, wspace=0.05)

# ---------------------------------------------------------
# Plot each month
# ---------------------------------------------------------
for i, month in enumerate(selected_months):
    ax = axes[i]

    zones["LST"] = lst_values[month]

    zones.plot(
        column="LST",
        cmap="RdYlGn_r",
        vmin=vmin,
        vmax=vmax,
        linewidth=0.3,
        edgecolor="black",
        ax=ax
    )

    # Hotspot marker and label
    ax.scatter(hot_pt.x, hot_pt.y, s=80, color="red")
    ax.text(hot_pt.x, hot_pt.y + 0.005,
            f"Hotspot\n({hot_name})", fontsize=7, color="red", ha="center")

    # Coolspot marker and label
    ax.scatter(cool_pt.x, cool_pt.y, s=80, color="blue")
    ax.text(cool_pt.x, cool_pt.y + 0.005,
            f"Coolspot\n({cool_name})", fontsize=7, color="blue", ha="center")

    ax.set_title(f"{month}", fontsize=12) # Removed "LST" from title
    ax.axis("off")

# ---------------------------------------------------------
# Unified Legend Panel
# ---------------------------------------------------------
legend_ax = fig.add_axes([0.1, 0.02, 0.8, 0.15])
legend_ax.axis("off")

common_y = 0.5

# North Arrow
na_ax = legend_ax.inset_axes([0.02, common_y - 0.25, 0.08, 0.5])
na_ax.axis("off")
arrow = FancyArrowPatch(
    posA=(0.5, 0.2),
    posB=(0.5, 0.9),
    mutation_scale=20,
    color='black',
    arrowstyle='-|>',
    linewidth=2,
    transform=na_ax.transAxes
)
na_ax.add_patch(arrow)
na_ax.text(0.5, 0.95, "N", ha="center", fontsize=12, fontweight="bold",
           transform=na_ax.transAxes)

# Colorbar
sm = plt.cm.ScalarMappable(cmap="RdYlGn_r", norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cax = legend_ax.inset_axes([0.2, common_y - 0.1, 0.4, 0.2])
cbar = fig.colorbar(sm, cax=cax, orientation="horizontal")
cbar.set_label("LST (°C)", fontsize=8)

# Scale Bar
scale_x = 0.75
scale_w = 0.12
legend_ax.plot([scale_x, scale_x + scale_w], [common_y, common_y],
               transform=legend_ax.transAxes, color="black", linewidth=4)
legend_ax.text(scale_x + scale_w / 2, common_y - 0.15, "2 km",
               ha="center", fontsize=10, transform=legend_ax.transAxes)

# ---------------------------------------------------------
# Hotspot/Coolspot Indicator Legend (bottom right)
# ---------------------------------------------------------
hotspot_patch = mpatches.Patch(color='red', label='Hotspot Zone')
coolspot_patch = mpatches.Patch(color='blue', label='Coolspot Zone')
hs_cs_handles = [hotspot_patch, coolspot_patch]

hs_cs_legend_ax = legend_ax.inset_axes([0.7, 0.05, 0.2, 0.4]) # [x, y, width, height] relative to legend_ax
hs_cs_legend_ax.axis('off')
hs_cs_legend_ax.legend(handles=hs_cs_handles, loc='center', fontsize=8, frameon=False)

# ---------------------------------------------------------
# Seasonal Annotations (Adjusted y-coordinates to move above maps)
# ---------------------------------------------------------
# Adjusted y-coordinates to be higher to avoid overlap with map plots
fig.text(0.2, 0.92,
         "March → April: Sharp increase in mean and maximum LST.\nPre-monsoon heating intensifies.",
         fontsize=9, ha='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.7))

fig.text(0.8, 0.92,
         "May → October: Drop in LST due to monsoon cooling.\nCooler zones reappear.",
         fontsize=9, ha='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.7))

# These annotations are for the second row of maps, keeping them around y=0.18-0.2 is appropriate
fig.text(0.32, 0.18,
         "October → November: Gradual cooling post-monsoon.\nThermal stress reduces.",
         fontsize=9, ha='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.7))

fig.text(0.7, 0.18,
         "November → December: Minimum LST drops significantly.\nWinter transition begins.",
         fontsize=9, ha='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=0.7))

# Title
fig.suptitle("Seasonal LST Variation (Significant Zones + Hotspot/Coolspot)",
             fontsize=16, weight="bold", y=0.98)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
from rasterstats import zonal_stats
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler # Import StandardScaler
import os

##"Monthly Heat-Risk Classification (K-means) – Dhaka, 2017##


# -----------------------------
# Paths
# -----------------------------
folder = "/content/drive/MyDrive/GEE_LST_Dhaka_2017"
shp_path = "/content/drive/MyDrive/ARS/Coredhaka.shp"

# Load zones (WGS84)
zones = gpd.read_file(shp_path)

# List all .tif files
files = sorted([f for f in os.listdir(folder) if f.endswith(".tif")])

month_names = ["Jan", "Feb", "Mar", "Apr",
               "May", "Jun", "Jul", "Aug",
               "Sep", "Oct", "Nov", "Dec"]

# Exclude cloudy months
exclude_months = ["Jun", "Jul", "Aug", "Sep"]

# Number of clusters
n_clusters = 4

zone_maps = []
zone_months = []

# -----------------------------
# Compute zonal LST + K-means clusters for each month
# -----------------------------
for fname, month in zip(files, month_names):

    if month in exclude_months:
        continue

    path = os.path.join(folder, fname)

    # Zonal LST
    stats = zonal_stats(
        zones,
        path,
        stats=["mean"],
        geojson_out=True,
        nodata=np.nan
    )

    gdf = gpd.GeoDataFrame.from_features(stats)
    gdf["LST"] = gdf["mean"].astype(float)

    # Drop NaN zones
    gdf = gdf.dropna(subset=["LST"])

    # Normalize LST
    gdf["LST_norm"] = (gdf["LST"] - gdf["LST"].min()) / (gdf["LST"].max() - gdf["LST"].min())

    # K-means clustering
    X = gdf[["LST_norm"]].values
    X_scaled = StandardScaler().fit_transform(X)

    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    gdf["Cluster"] = kmeans.fit_predict(X_scaled)

    zone_maps.append(gdf)
    zone_months.append(month)

# -----------------------------
# Plot monthly cluster maps
# -----------------------------
n = len(zone_maps)
cols = 4
rows = int(np.ceil(n / cols))

# Remove constrained_layout=True to manually adjust spacing
fig, axes = plt.subplots(rows, cols, figsize=(16, 4.5 * rows))
axes = np.array(axes).reshape(-1)

# Adjust subplot parameters to make space at the bottom for legends
# Increased bottom margin to 0.2 to leave more space for the unified legend panel
fig.subplots_adjust(bottom=0.2, top=0.9, left=0.05, right=0.95,
                    hspace=0.1, wspace=0.05)

used_axes = axes[:n]

for ax, gdf, month in zip(used_axes, zone_maps, zone_months):
    gdf.plot(
        column="Cluster",
        cmap="Set1",
        linewidth=0.3,
        edgecolor="black",
        legend=False,
        ax=ax
    )
    ax.set_title(month, fontsize=12, weight="bold")
    ax.axis("off")

# Hide unused axes
for ax in axes[n:]:
    ax.axis("off")

# ---------------------------------------------------------
# Unified Legend Panel at the bottom
# ---------------------------------------------------------
# Create a single Axes to host all legend elements at the bottom
# Adjusted bottom to be very low and height to be compact (e.g., 0.08-0.1)
legend_ax = fig.add_axes([0.1, 0.02, 0.8, 0.1]) # [left, bottom, width, height]
legend_ax.axis("off")

# Define a common vertical center for elements in legend_ax for alignment
common_y_pos = 0.5

# Heat-Risk Category Legend
from matplotlib.lines import Line2D
risk_labels = [
    "Low Risk",
    "Lower-Medium Risk",
    "Medium Risk",
    "High Risk"
]
colors = plt.cm.Set1(np.linspace(0, 1, n_clusters))
handles = [
    Line2D(
        [0], [0],
        marker='s',
        color='none',
        markerfacecolor=colors[i],
        markersize=8,
        label=risk_labels[i]
    )
    for i in range(n_clusters)
]
# Place the legend inside the custom legend_ax
# Adjusted bbox_to_anchor to position it centrally and vertically aligned
legend_ax.legend(
    handles=handles,
    loc="center", # Use center for consistent placement
    ncol=n_clusters,
    fontsize=10,
    title="Heat-Risk Category",
    bbox_to_anchor=(0.5, common_y_pos + 0.1), # Align vertically slightly above the center line
    frameon=False
)

# North Arrow
# Using an inset_axes for precise positioning within the legend_ax
# Adjusted vertical position to align with other elements' centers
n_arrow_inset_ax = legend_ax.inset_axes([0.05, common_y_pos - 0.2, 0.08, 0.6]) # [x, y, width, height]
n_arrow_inset_ax.axis('off')
n_arrow_inset_ax.annotate('N', xy=(0.5, 0.85), xytext=(0.5, 0.15),
                         ha="center", va="center", fontsize=14, fontweight="bold",
                         arrowprops=dict(facecolor="black", width=3, headwidth=10),
                         xycoords="axes fraction", textcoords="axes fraction")


# Scale Bar
# Positioned to the right within the legend_ax
scale_bar_x_start = 0.8 # Relative to legend_ax
scale_bar_width_relative = 0.12 # Relative width

# Line for scale bar - aligned vertically
legend_ax.plot(
    [scale_bar_x_start, scale_bar_x_start + scale_bar_width_relative],
    [common_y_pos + 0.1, common_y_pos + 0.1], # Align vertically with text/legend middle
    transform=legend_ax.transAxes,
    color="black",
    linewidth=3
)

# Text for scale bar - aligned vertically
legend_ax.text(
    scale_bar_x_start + scale_bar_width_relative / 2,
    common_y_pos - 0.2, # Position text below the line
    "2 km",
    ha="center",
    va="top",
    fontsize=10,
    transform=legend_ax.transAxes
)

# -----------------------------
# Title
# -----------------------------
fig.suptitle(
    "Monthly Heat-Risk Classification (K-means) – Dhaka, 2017\n",
    fontsize=16, weight="bold",
    y=0.98 # Adjust title position if needed
)

plt.show()
```

```python
def gdf_to_ee(gdf):
    features = []
    for _, row in gdf.iterrows():
        geom = row.geometry

        if geom.geom_type == "Polygon":
            ee_geom = ee.Geometry.Polygon(list(geom.exterior.coords))

        elif geom.geom_type == "MultiPolygon":
            polygons = []
            for poly in geom.geoms:
                polygons.append(list(poly.exterior.coords))
            ee_geom = ee.Geometry.MultiPolygon(polygons)

        else:
            raise ValueError(f"Unsupported geometry type: {geom.geom_type}")

        feat = ee.Feature(ee_geom, {"NAME_3": row["NAME_3"]})
        features.append(feat)

    return ee.FeatureCollection(features)

zones = gdf_to_ee(dhaka_zones)
print("Number of zones in EE:", zones.size().getInfo())
```

```python
lst_ic = ee.ImageCollection("MODIS/061/MOD11A2").select("LST_Day_1km")
```

```python
def compute_year(year):
    start = ee.Date.fromYMD(year, 1, 1)
    end   = ee.Date.fromYMD(year, 12, 31)

    col = lst_ic.filterDate(start, end)

    # Skip empty collections
    if col.size().getInfo() == 0:
        print(f"WARNING: No MODIS data for {year}")
        return ee.FeatureCollection([])

    img = (
        col.mean()
           .multiply(0.02)      # scale factor
           .subtract(273.15)    # Kelvin → Celsius
           .rename("LST")
    )

    stats = img.reduceRegions(
        collection=zones,
        reducer=ee.Reducer.mean(),
        scale=1000
    ).map(lambda f: f.set("year", year))

    return stats
```

```python
import pandas as pd
import numpy as np
import pymannkendall as mk
from scipy.stats import theilslopes

# ---------------------------------------------------------
# INPUT REQUIREMENT:
# Using df_modis_yearly (from previous steps) with columns:
#   NAME_3 | year | mean
# ---------------------------------------------------------

df_to_analyze = df_modis_yearly.copy() # Use the correct DataFrame

zones = df_to_analyze['NAME_3'].unique()
results = []

for z in zones:
    subset = df_to_analyze[df_to_analyze['NAME_3'] == z].sort_values('year')
    years = subset['year'].values
    lst_values = subset['mean'].values.astype(float) # Ensure LST values are float

    # Skip zones with insufficient data (e.g., less than 5 years for robust trend)
    if len(lst_values) < 5:
        results.append({
            'Zone': z,
            'MK_Z': np.nan,
            'MK_p': np.nan,
            'MK_Trend': 'Insufficient data',
            'TS_Slope_per_year': np.nan,
            'TS_Lower_CI': np.nan,
            'TS_Upper_CI': np.nan
        })
        continue

    # Mann–Kendall Test
    mk_test = mk.original_test(lst_values)

    # Theil–Sen Slope
    # theilslopes expects x, y, not y, x. Years are x, LST values are y.
    slope, intercept, lo_slope, hi_slope = theilslopes(lst_values, years)

    results.append({
        'Zone': z,
        'MK_Z': mk_test.z,
        'MK_p': mk_test.p,
        'MK_Trend': mk_test.trend,
        'TS_Slope_per_year': slope,
        'TS_Lower_CI': lo_slope,
        'TS_Upper_CI': hi_slope
    })

# Convert to DataFrame
trend_df = pd.DataFrame(results)

# Sort by slope (optional)
trend_df = trend_df.sort_values('TS_Slope_per_year', ascending=False)

# Print full table
pd.set_option('display.max_rows', None)
print(trend_df)
```

```python
import numpy as np
from scipy import stats
import pandas as pd

def calculate_lst_trend(df):
    """
    Calculate linear trend for LST time series
    """
    # Ensure proper datetime format
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])

    # Convert dates to numeric (days since first observation)
    df['days_since_start'] = (df['Timestamp'] - df['Timestamp'].min()).dt.days

    # Remove any rows with missing LST values
    df_clean = df.dropna(subset=['MeanLST']).copy()

    if len(df_clean) < 2:
        print("ERROR: Not enough valid data points for trend analysis!")
        return

    # Linear regression: days vs MeanLST
    x = df_clean['days_since_start'].values
    y = df_clean['MeanLST'].values

    # Calculate regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

    # Convert slope to meaningful units
    slope_per_day = slope
    slope_per_year = slope * 365.25
    total_years = (df_clean['Timestamp'].max() - df_clean['Timestamp'].min()).days / 365.25
    total_change = slope_per_year * total_years

    # Output results
    print("=== DECADAL TREND ANALYSIS ===")
    print(f"Time period: {df_clean['Timestamp'].min().date()} to {df_clean['Timestamp'].max().date()}")
    print(f"Valid data points: {len(df_clean)} of {len(df)}")
    print(f"Slope: {slope_per_day:.6f}°C per day")
    print(f"Annual trend: {slope_per_year:.3f}°C per year")
    print(f"Total change ({total_years:.1f} years): {total_change:.3f}°C")
    print(f"R-squared: {r_value**2:.3f}")
    print(f"P-value: {p_value:.6f}")

    # Statistical significance (95% confidence level)
    significant = p_value < 0.05
    print(f"Significant trend (p<0.05): {'YES' if significant else 'NO'}")

    # Interpretation
    if significant:
        direction = "increasing" if slope_per_year > 0 else "decreasing"
        print(f"\nInterpretation: Statistically significant {direction} trend of {abs(slope_per_year):.3f}°C per year")
    else:
        print("\nInterpretation: No statistically significant trend detected")

    return {
        'slope_per_day': slope_per_day,
        'slope_per_year': slope_per_year,
        'total_change': total_change,
        'r_squared': r_value**2,
        'p_value': p_value,
        'significant': significant,
        'n_points': len(df_clean),
        'intercept': intercept # Add intercept here
    }

# Run the analysis
results = calculate_lst_trend(df)
```

```python
import matplotlib.pyplot as plt

def plot_lst_trend(df, results):
    """Visualize LST time series with trend line"""

    # First, ensure we have the same data cleaning as in calculate_lst_trend
    df_plot = df.copy()
    df_plot['Timestamp'] = pd.to_datetime(df_plot['Timestamp'])
    df_plot['days_since_start'] = (df_plot['Timestamp'] - df_plot['Timestamp'].min()).dt.days
    df_plot = df_plot.dropna(subset=['MeanLST'])

    fig, ax = plt.subplots(figsize=(12, 6))

    # Scatter plot of actual data
    ax.scatter(df_plot['Timestamp'], df_plot['MeanLST'],
               alpha=0.6, s=30, label='LST Observations', color='blue')

    # Trend line (FIXED: using results['intercept'])
    if results and results['n_points'] > 0:
        x_dates = df_plot['Timestamp'].values
        x_numeric = (x_dates - x_dates.min()) / np.timedelta64(1, 'D')
        trend_line = results['slope_per_day'] * x_numeric + results['intercept']  # ← Fixed!

        ax.plot(df_plot['Timestamp'], trend_line,
                'r-', linewidth=3,
                label=f"Trend: {results['slope_per_year']:+.3f}°C/yr (p={results['p_value']:.3f})")

    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Land Surface Temperature (°C)', fontsize=12)

    # Create title with period info
    start_year = df_plot['Timestamp'].min().year
    end_year = df_plot['Timestamp'].max().year
    title = f'LST Time Series Trend ({start_year}-{end_year})'

    # Add significance info to title
    if results['significant']:
        direction = "↑ Increasing" if results['slope_per_year'] > 0 else "↓ Decreasing"
        title += f'\n{direction}: {abs(results["slope_per_year"]):.3f}°C/yr (p<0.05)'
    else:
        title += '\nNo significant trend (p>0.05)'

    ax.set_title(title, fontsize=14)
    ax.legend(fontsize=11)
    ax.grid(True, alpha=0.3)

    # Add statistics box (adjusted position)
    stats_text = f"n = {results['n_points']}\nR² = {results['r_squared']:.3f}"
    ax.text(0.98, 0.5, stats_text, transform=ax.transAxes,
            fontsize=10, verticalalignment='center', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Create the visualization
if results:
    plot_lst_trend(df, results)
```

```python
### NDVI trend map


import geopandas as gpd
import pandas as pd
import numpy as np
from scipy.stats import mstats
import matplotlib.pyplot as plt
import rasterio
import rasterio.mask # Explicitly import the mask submodule
import os

# --- Helper function to extract zonal means (re-using from ed6727f4) ---
def extract_zone_means(raster_path, zones_gdf):
    with rasterio.open(raster_path) as src:
        results = []
        for idx, row in zones_gdf.iterrows():
            geom = [row.geometry.__geo_interface__]
            try:
                out_image, _ = rasterio.mask.mask(src, geom, crop=True)
                data = out_image[0].astype(float)
                data[data == src.nodata] = np.nan
                mean_val = np.nanmean(data)
            except Exception as e:
                mean_val = np.nan
            results.append({"NAME_3": row["NAME_3"], "value": mean_val})
    return pd.DataFrame(results)

# --- Load dhaka_zones (GeoDataFrame) ---
dhaka_zones = gpd.read_file("/content/drive/MyDrive/ARS/Coredhaka.shp")

# ============================================
# 1. Prepare NDVI Data
# ============================================
# gdf_ndvi: GeoDataFrame with columns:
# NAME_3, year, NDVI_mean, geometry

# Create ndvi_zone_mean (re-using logic from 31572374)
ndvi_base = "/content/drive/MyDrive/GEE_NDVI_Dhaka_2014_2024"
ndvi_records = []

for year in range(2014, 2025):
    ndvi_path = f"{ndvi_base}/NDVI_{year}.tif"

    if not os.path.exists(ndvi_path):
        print(f"Missing NDVI for {year}")
        continue

    df_temp = extract_zone_means(ndvi_path, dhaka_zones) # Pass the GeoDataFrame here
    df_temp["year"] = year
    ndvi_records.append(df_temp)

ndvi_df = pd.concat(ndvi_records)

ndvi_zone_mean = (
    ndvi_df.groupby("NAME_3")["value"]
           .mean()
           .reset_index()
           .rename(columns={"value": "NDVI"})
)

# Create gdf_ndvi by merging existing data
zones_gdf = dhaka_zones.copy() # Ensure zones_gdf is a GeoDataFrame
gdf_ndvi = zones_gdf.merge(ndvi_zone_mean, on="NAME_3", how="left")
gdf_ndvi['NDVI_mean'] = gdf_ndvi['NDVI'] # Rename 'NDVI' column to 'NDVI_mean'

# Filter out zones with NaN NDVI values before further processing
gdf_ndvi = gdf_ndvi.dropna(subset=['NDVI_mean'])

# Ensure 'year' is present and correct in gdf_ndvi for trend calculation
# If NDVI_mean is an overall average, we need yearly data for mstats.theilslopes
# Assuming you want to calculate trend across the 2014-2024 period, you need yearly NDVI means for each zone
# The current ndvi_zone_mean is an average across all years, so it cannot be used directly for yearly trends.
# We need to re-structure gdf_ndvi to contain yearly NDVI means per zone

# Re-create gdf_ndvi with yearly data for trend calculation
yearly_ndvi_data = []
for zone in dhaka_zones["NAME_3"].unique():
    zone_subset = ndvi_df[ndvi_df["NAME_3"] == zone].copy()
    for year in range(2014, 2025):
        if year in zone_subset['year'].values:
            yearly_ndvi_data.append({
                'NAME_3': zone,
                'year': year,
                'NDVI_mean': zone_subset[zone_subset['year'] == year]['value'].mean() # Get yearly mean
            })

if yearly_ndvi_data:
    gdf_ndvi_yearly = pd.DataFrame(yearly_ndvi_data)
    # Merge with geometry
    gdf_ndvi = dhaka_zones.merge(gdf_ndvi_yearly, on='NAME_3', how='left')
    gdf_ndvi = gdf_ndvi.dropna(subset=['NDVI_mean']) # Drop zones with no data for any year
    gdf_ndvi = gdf_ndvi.set_geometry("geometry")
else:
    print("Warning: No yearly NDVI data collected for trend analysis.")
    gdf_ndvi = dhaka_zones.copy() # Fallback to empty GeoDataFrame to prevent further errors
    gdf_ndvi['NDVI_mean'] = np.nan
    gdf_ndvi['year'] = np.nan


zones = gdf_ndvi["NAME_3"].unique()
years_for_trend = sorted(gdf_ndvi["year"].unique())

# Create GeoDataFrame to store trend results
ndvi_trend = gdf_ndvi.drop_duplicates(subset=["NAME_3"])[["NAME_3", "geometry"]].copy()
ndvi_trend["sen_slope"] = np.nan
ndvi_trend["intercept"] = np.nan

# ============================================
# 2. Compute Sen's slope per zone
# ============================================
for zone in zones:
    # Filter the yearly data for the current zone
    ts = gdf_ndvi[gdf_ndvi["NAME_3"] == zone].sort_values('year')

    if len(ts) < 2:
        continue
    x = ts["year"].values
    y = ts["NDVI_mean"].values.astype(float)

    # Handle cases where all y values are NaN for a zone
    if np.all(np.isnan(y)):
        continue

    # theilslopes can handle NaN if there are enough non-NaN values
    valid_indices = ~np.isnan(y)
    if np.sum(valid_indices) < 2: # Need at least 2 valid points for slope
        continue

    slope, intercept, lower, upper = mstats.theilslopes(y[valid_indices], x[valid_indices], 0.95)
    ndvi_trend.loc[ndvi_trend["NAME_3"] == zone, "sen_slope"] = slope
    ndvi_trend.loc[ndvi_trend["NAME_3"] == zone, "intercept"] = intercept

# ============================================
# 3. Plot NDVI Trend Map
# ============================================
fig, ax = plt.subplots(figsize=(12,10))

# Blue = greening, brown = vegetation loss
ndvi_trend.plot(
    column="sen_slope",
    cmap="BrBG",
    edgecolor="black",
    linewidth=0.4,
    legend=True,
    legend_kwds={
        "label": "NDVI Sen's Slope (per year)",
        "shrink": 0.6
    },
    ax=ax
)

ax.set_title("NDVI Trend (Sen's Slope) per Zone", fontsize=16)
ax.axis("off")


# ---------------------------------------------------------
# NORTH ARROW (center-left)
# ---------------------------------------------------------
na_ax = fig.add_axes([0.05, 0.50, 0.08, 0.20])  # [left, bottom, width, height]
na_ax.axis("off")

arrow = FancyArrowPatch(
    posA=(0.5, 0.1),
    posB=(0.5, 0.9),
    arrowstyle='-|>',
    mutation_scale=28,
    color='black',
    linewidth=2,
    transform=na_ax.transAxes
)
na_ax.add_patch(arrow)

na_ax.text(
    0.5, 0.95,
    "N",
    ha="center",
    va="center",
    fontsize=14,
    fontweight="bold",
    transform=na_ax.transAxes
)

# ---------------------------------------------------------
# SCALE BAR (center-left, below arrow)
# ---------------------------------------------------------
meters_per_degree_lat = 110574  # Dhaka latitude ~23.8°
deg_for_5km = 5000 / meters_per_degree_lat

sb_ax = fig.add_axes([0.05, 0.28, 0.15, 0.08])  # Adjusted bottom from 0.33 to 0.28
sb_ax.axis("off")

sb_ax.plot(
    [0.1, 0.9],
    [0.5, 0.5],
    color="black",
    linewidth=4,
    transform=sb_ax.transAxes
)

sb_ax.text(
    0.5, 0.0,
    "5 km",
    ha="center",
    va="bottom",
    fontsize=10,
    transform=sb_ax.transAxes
)
# Optional: overlay slum boundaries
if 'gdf_slums' in globals():
    gdf_slums.boundary.plot(ax=ax, color="black", linewidth=1, alpha=0.5)

plt.show()
```
